```python
# ============================================
# 🧠 YOLOv11 객체 탐지 실습 (YouTube 영상 사용)
# ============================================

# 1️⃣ 기본 패키지 설치
!pip install ultralytics yt_dlp moviepy -q

# 2️⃣ 라이브러리 불러오기
from ultralytics import YOLO
from moviepy.editor import VideoFileClip
import os

# 3️⃣ 유튜브 영상 다운로드 (yt_dlp 사용)
yt_url = "https://www.youtube.com/shorts/9qzj22IzBeE"  # 원하는 영상 링크
os.makedirs("videos", exist_ok=True)
video_path = "videos/input_video.mp4"

# yt-dlp로 다운로드 (최고 화질)
!yt-dlp -f "best[ext=mp4]" -o {video_path} {yt_url}

# 4️⃣ 15초짜리로 자르기
full_clip = VideoFileClip(video_path)
video_duration = full_clip.duration
full_clip.close()  # 파일을 닫아 메모리 해제

# 15초까지 자르기, 영상이 15초보다 짧으면 끝까지 자름
end_time = min(15, video_duration)
clip = VideoFileClip(video_path).subclip(0, end_time)
clip.write_videofile("videos/input_clip.mp4", codec="libx264", audio_codec="aac")

# 5️⃣ YOLOv11 모델 불러오기 (자동 다운로드됨)
model = YOLO("yolo11n.pt")  # YOLOv11n 모델 사용 (경량 모델)

# 6️⃣ 객체 탐지 실행
results = model.predict(
    source="videos/input_clip.mp4",  # 입력 영상
    save=True,                       # 결과 영상 저장
    project="runs/detect",           # 저장 폴더
    name="yt_demo",                  # 세부 폴더명
    conf=0.4                         # confidence threshold
)

# 7️⃣ 결과 확인 (Colab에서 직접 재생)
from IPython.display import Video, display
display(Video("runs/detect/yt_demo/input_clip.mp4", embed=True))
```
